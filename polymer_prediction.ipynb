```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
train_data["SMILES"] = LE.fit_transform(train_data["SMILES"])
test_data["SMILES"] = LE.fit_transform(test_data["SMILES"])      #to convert text format of smiles to number, making it suitable for ML
```

```python
train_data.head()
```

```python
# machine learning:
```

```python
train_ffv = train_data.dropna(subset=["FFV"])    
# Removes rows where the FFV value is missing, so the model can be trained using complete dat
```

```python
train_ffv.head()
train_ffv.shape
```

```python
x=train_ffv[["SMILES"]]
y=train_ffv["FFV"]
#x has smiles, y has FFV which is to be predicted
```

```python
print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train, y_train)     #creates a LR model and trains it using dataset
```

```python
y_pred=LR.predict(x_test)
print(y_pred[:10])
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
test_predictions=LR.predict(test_data[["SMILES"]])
print(test_predictions[:10])
```
