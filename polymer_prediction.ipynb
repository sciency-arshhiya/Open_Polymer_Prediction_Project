```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
train_df["SMILES"] = LE.fit_transform(train_df["SMILES"])
test_df["SMILES"] = LE.fit_transform(test_df["SMILES"])      #to convert text format of smiles to number, making it suitable for ML
```

```python
# machine learning:
```

```python
train_ffv = train_df.dropna(subset=["FFV"])    
# Removes rows where the FFV value is missing, so the model can be trained using complete dat
```

```python
x=train_ffv[["SMILES"]]
y=train_ffv["FFV"]
#x has smiles, y has FFV which is to be predicted
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train, y_train)     #creates a LR model and trains it using dataset
```

```python
y_pred=LR.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
